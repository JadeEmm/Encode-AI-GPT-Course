# 2. Run sentiment analysis predictions by using Pipeline

```python
# Using pipeline class to make predictions from models available in the Hub
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
data = ["In the middle",
        "I have doubts",
        "I'm very pessimistic and I'm very optimistic",
        "I'm very pessimistic",
        "I'm very optimistic"]
sentiment_pipeline(data)
```

# 3. Use a specific sentiment analysis model available on the Hub by specifying its name

```python
# Using bertweet model for sentiment analysis
bertweet_model = pipeline(model="finiteautomata/bertweet-base-sentiment-analysis")
bertweet_model(data)
```

```python
# Using cardiffnlp model for sentiment analysis
roberta_model = pipeline(model="cardiffnlp/twitter-roberta-base-sentiment-latest")
roberta_model(data)
```
